# Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import qgrid
```

# Get Data 

```python
df = pd.read_excel("C:\\Users\\gfulham\\Code\\Projects\\survey-monkey\\data\\survey_monkey_final.xlsx")
widget = qgrid.show_grid(df)
widget
```

# Transform Data

```python
# total shows the total count each answer 1-8 recieved from all the questions
total = df.groupby(["Answer"]).count()["Respondents"]
total
```

# Visualize Data

```python
fig = plt.figure(figsize = (16,8))
st = fig.suptitle("Survey Answer Order Bias", fontsize = 'xx-large')

ax1 = fig.add_subplot(121)
ax1.plot(total)
plt.ylabel("Total Responses", fontsize="large")
plt.ylim(bottom = 1)

ax2 = fig.add_subplot(122)
plt.ylabel("Total Responses", fontsize="large")
ax2.plot(total)
plt.savefig("Survey Answer Order Bias")
```

# Conclusion

In this single survey there seems to be a small amount of bias to answering the questions with answers 4-7. If the Survey is designed on a Likert scale, the normal distribution might be off, looking at the low amount of total responses for answer 3. This distribution of responses may be acceptable due to the number of respondents(198). Depends on the industry standard for question and answer bias.
